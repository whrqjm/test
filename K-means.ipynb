```python
x = np.random.rand(400)
y = np.random.rand(400)
data = np.concatenate((x.reshape(len(x), 1), y.reshape(len(y), 1)), axis = 1)

epoch = 0
k = 4
while True:
    epoch += 1
    print("{0}回目".format(epoch))

    clusters_init = np.random.randint(0, k, (400, 1))

    data = np.concatenate((data, clusters_init), axis = 1)

    ccx = [np.mean(data[:, 0][np.where(data[:, 2] == i)]) for i in range(k)]
    ccy = [np.mean(data[:, 1][np.where(data[:, 2] == i)]) for i in range(k)]
    
    nc = [
        np.argmin(
            [np.linalg.norm(data[i, :2] - [ccx[j], ccy[j]]) for j in range(k)]
        ) for i in range(len(data))
    ]
    
    if np.array_equal(nc, data[:, 2]):
        print("おしまい")
        break
    else:
        data[:, 2] = nc
        
plt.scatter(data[np.where(data[:, 2] == 0)][:, 0], data[np.where(data[:, 2] == 0)][:, 1], color = "red")
plt.scatter(data[np.where(data[:, 2] == 1)][:, 0], data[np.where(data[:, 2] == 1)][:, 1], color = "blue")
plt.scatter(data[np.where(data[:, 2] == 2)][:, 0], data[np.where(data[:, 2] == 2)][:, 1], color = "green")
plt.scatter(data[np.where(data[:, 2] == 3)][:, 0], data[np.where(data[:, 2] == 3)][:, 1], color = "yellow")
plt.show()
```
